## üì¶ Import Libraries and Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from datetime import datetime, timedelta
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('default')
plt.rcParams['figure.figsize'] = (20, 16)
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['figure.facecolor'] = 'white'

# Import modular scripts
import sys
sys.path.append('.')
from scripts.stock_data_loader import load_stock_data_from_db
from scripts.fibonacci_pivot_detector import (
    detect_fibonacci_pivots,
    create_fibonacci_swings,
    plot_fibonacci_pivots,
    analyze_fibonacci_pivots,
    validate_pivot_sequence
)

print("‚úÖ Libraries and modular scripts imported successfully!")
print("üìà Ready for maximum-sensitivity Fibonacci pivot analysis")
print("üîß Using modular fibonacci_pivot_detector.py for clean, reusable code")
```

## ‚öôÔ∏è Configuration

```python
# Analysis configuration
SYMBOL = 'QQQ'  # Target symbol for analysis
ANALYSIS_DAYS = 365  # One calendar year of data
TIMEFRAME = '1D'  # Daily data

# Maximum-sensitivity Fibonacci pivot detection parameters
FIBONACCI_PARAMS = {
    'lookback_window': 5,        # Maximum sensitivity: 5-day lookback
    'min_strength': 0.0005,      # Maximum sensitivity: 0.05% threshold
    'trend_confirmation': 1      # Minimal confirmation: 1 day
}

# Chart configuration
CHART_CONFIG = {
    'figsize': (20, 16),        # Large size for clarity
    'enhanced_labels': True      # Crystal clear labeling
}

print(f"üéØ Fibonacci Analysis Configuration:")
print(f"   Symbol: {SYMBOL}")
print(f"   Analysis Period: {ANALYSIS_DAYS} days")
print(f"   Timeframe: {TIMEFRAME}")
print(f"\nüîß Maximum-Sensitivity Parameters:")
print(f"   Lookback window: {FIBONACCI_PARAMS['lookback_window']} days")
print(f"   Strength threshold: {FIBONACCI_PARAMS['min_strength']*100:.3f}%")
print(f"   Trend confirmation: {FIBONACCI_PARAMS['trend_confirmation']} day(s)")
print(f"\nüìä Chart Configuration:")
print(f"   Figure size: {CHART_CONFIG['figsize']}")
print(f"   Enhanced labeling: {CHART_CONFIG['enhanced_labels']}")
```

## üìä Load Stock Data

```python
# Load stock data
print(f"üìä Loading {ANALYSIS_DAYS} days of {SYMBOL} data...")

stock_data = load_stock_data_from_db(
    symbol=SYMBOL,
    days=ANALYSIS_DAYS,
    timeframe=TIMEFRAME,
    filter_premarket=True
)

if len(stock_data) > 0:
    # Ensure LogPrice column exists
    if 'LogPrice' not in stock_data.columns:
        stock_data['LogPrice'] = np.log(stock_data['Price'])
        print("   Added LogPrice column for analysis")

    print(f"‚úÖ Loaded {len(stock_data)} trading days")
    print(f"   Date range: {stock_data['Date'].min().date()} to {stock_data['Date'].max().date()}")
    print(f"   Price range: ${stock_data['Price'].min():.2f} - ${stock_data['Price'].max():.2f}")
    print(f"   Current price: ${stock_data['Price'].iloc[-1]:.2f}")

    # Display sample data
    print(f"\nüìã Sample Data (First 5 rows):")
    display(stock_data[['Date', 'Price', 'Volume', 'LogPrice']].head())
else:
    print(f"‚ùå No stock data loaded! Please check data availability.")
    raise ValueError("No stock data available")
```

## üéØ Complete Fibonacci Analysis Using Module

```python
# Perform complete Fibonacci pivot analysis using the modular approach
print(f"üéØ Performing MAXIMUM-SENSITIVITY Fibonacci Analysis for {SYMBOL}...")
print(f"   üöÄ Using modular fibonacci_pivot_detector.py")
print(f"   üìä Capturing EVERY significant turning point with perfect alternating sequences")

# Run complete analysis
analysis_results = analyze_fibonacci_pivots(stock_data, **FIBONACCI_PARAMS)

# Extract results
fibonacci_pivots = analysis_results['pivots']
fibonacci_swings = analysis_results['swings']
validation_results = analysis_results['validation']
statistics = analysis_results['statistics']

# Display comprehensive results
print(f"\n‚úÖ MAXIMUM-SENSITIVITY Fibonacci Analysis Complete!")
print(f"\nüìä Detection Statistics:")
print(f"   Total Fibonacci pivots: {statistics['total_pivots']}")
print(f"   Fibonacci highs: {statistics['highs']}")
print(f"   Fibonacci lows: {statistics['lows']}")
print(f"   Total swings: {statistics['total_swings']}")
if statistics['total_pivots'] > 0:
    print(f"   Average pivot strength: {statistics['avg_strength']*100:.3f}%")
    print(f"   Strength range: {statistics['strength_range'][0]*100:.3f}% - {statistics['strength_range'][1]*100:.2f}%")

# Display validation results
print(f"\n‚úÖ Sequence Validation:")
print(f"   Perfect alternating sequence: {'‚úÖ YES' if validation_results['alternating'] else '‚ùå NO'}")
print(f"   Price rule validation: {validation_results['validation_rate']:.1f}% pass rate")
if validation_results['violations'] > 0:
    print(f"   Price violations: {validation_results['violations']} (kept for edge cases)")
print(f"   Alternating sequence: {validation_results['sequence'][:30]}{'...' if len(validation_results['sequence']) > 30 else ''}")

# Show sample pivots
if fibonacci_pivots:
    print(f"\nüìã Sample Fibonacci Pivots (First 15):")
    for i, pivot in enumerate(fibonacci_pivots[:15]):
        pivot_date = pd.to_datetime(pivot['date']).date()
        print(f"   P{i+1:2d}: {pivot_date} ${pivot['price']:7.2f} ({pivot['type']:4s}) - {pivot['strength']*100:6.3f}%")

    if len(fibonacci_pivots) > 15:
        print(f"   ... and {len(fibonacci_pivots) - 15} more pivots")

# Show sample swings
if fibonacci_swings:
    print(f"\nüåÄ Sample Fibonacci Swings (First 10):")
    for swing in fibonacci_swings[:10]:
        start_date = pd.to_datetime(swing['start_date']).date()
        end_date = pd.to_datetime(swing['end_date']).date()
        print(f"   S{swing['swing_id']:2d}: {swing['swing_type']:9s} | {start_date} ‚Üí {end_date} | "
              f"{swing['percentage_move']:+6.1f}% | {swing['duration_days']:2d}d")

print(f"\nüéØ MODULAR FIBONACCI ANALYSIS SUCCESS!")
print(f"   ‚Ä¢ Clean, reusable code using fibonacci_pivot_detector.py module")
print(f"   ‚Ä¢ Maximum sensitivity with {statistics['total_pivots']} pivots detected")
print(f"   ‚Ä¢ Perfect alternating sequences for accurate Fibonacci calculations")
print(f"   ‚Ä¢ Ready for professional chart visualization")
```

## üìä Professional Chart Visualization

```python
# Generate professional Fibonacci pivot chart using the module
if fibonacci_pivots:
    print(f"üìä Generating professional Fibonacci pivot chart using modular visualization...")
    print(f"\nüéØ Chart Features:")
    print(f"   ‚Ä¢ Crystal clear labeling with enhanced contrast")
    print(f"   ‚Ä¢ Professional styling with larger markers and better visibility")
    print(f"   ‚Ä¢ Comprehensive information display")
    print(f"   ‚Ä¢ Volume correlation highlighting")
    print(f"   ‚Ä¢ Perfect readability for all {len(fibonacci_pivots)} pivots")

    # Create professional chart using the module
    fig = plot_fibonacci_pivots(
        stock_data=stock_data,
        fibonacci_pivots=fibonacci_pivots,
        fibonacci_swings=fibonacci_swings,
        symbol=SYMBOL,
        figsize=CHART_CONFIG['figsize']
    )

    plt.show()

    print(f"\n‚úÖ Professional chart generated successfully!")
    print(f"   üöÄ VISUALIZATION HIGHLIGHTS:")
    print(f"      ‚Ä¢ All {len(fibonacci_pivots)} pivots clearly labeled and numbered")
    print(f"      ‚Ä¢ {len(fibonacci_swings)} swing pairs with percentage moves and duration")
    print(f"      ‚Ä¢ Perfect alternating {validation_results['sequence']} sequence")
    print(f"      ‚Ä¢ Enhanced readability with professional styling")
    print(f"      ‚Ä¢ Complete volume correlation analysis")
    print(f"      ‚Ä¢ Ready for Fibonacci retracement and extension analysis")

else:
    print(f"‚ùå No Fibonacci pivots detected for chart generation")
```

## üìö Module Documentation

### Core Functions from `fibonacci_pivot_detector.py`:

**`detect_fibonacci_pivots(stock_data, lookback_window=5, min_strength=0.0005, trend_confirmation=1)`**
- Maximum-sensitivity pivot detection with alternating sequence validation
- Returns list of pivot dictionaries with date, price, type, strength, and validation

**`create_fibonacci_swings(fibonacci_pivots)`**
- Creates swing analysis from detected pivots for trend analysis
- Returns list of swing dictionaries with metrics and Fibonacci-ready data

**`plot_fibonacci_pivots(stock_data, fibonacci_pivots, fibonacci_swings, symbol, figsize)`**
- Professional chart visualization with crystal clear labeling
- Enhanced styling, larger markers, and better readability

**`analyze_fibonacci_pivots(stock_data, **kwargs)`**
- Complete analysis workflow combining detection, swings, and validation
- Returns comprehensive results dictionary with all analysis components

**`validate_pivot_sequence(fibonacci_pivots, tolerance=0.002)`**
- Validates alternating sequence and price rules with configurable tolerance
- Returns validation statistics and quality metrics

### Key Parameters:
- **lookback_window**: 5 days (maximum sensitivity)
- **min_strength**: 0.05% (captures smallest significant moves)
- **trend_confirmation**: 1 day (minimal confirmation requirement)
- **tolerance**: 0.2% (price validation tolerance for edge cases)

### Benefits of Modular Architecture:
1. **Clean Code Separation**: Reusable functions in dedicated module
2. **Professional Visualization**: Enhanced charts with crystal clear labeling
3. **Complete Workflow**: Single function for full analysis pipeline
4. **Validation System**: Built-in quality assurance and metrics
5. **Flexibility**: Configurable parameters for different analysis needs
6. **Maintainability**: Easy to update and extend functionality
